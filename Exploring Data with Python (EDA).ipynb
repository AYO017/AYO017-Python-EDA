```python
## Rotatethe ticklabels
plt.figure(figsize=(15,6)) # this creats a figure 8 inch wide, 4 inch high
ax = sns.countplot(x = "BusinessType", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40, ha="right")
#plt.show()
plt.show()
```
